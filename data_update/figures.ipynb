```python
import pandas as pd
import os
import sys
import dotenv

sys.path.append(os.getcwd())

dotenv.load_dotenv(".env")
data_dir = os.getenv("DATA_PATH")
# data_dir = "G:\GITAR\PubData"


# set data dir to "G:\GITAR\PubData"
# data_dir = r"G:\GITAR\PubData\\"

# Bring in new species lists

gbif_backbone = pd.read_csv( data_dir + "\GBIF data\GBIF_backbone_invasive.csv", dtype={"usageKey": str})
```

```python
#from backbone, make barplot of unique phyla. If phylum has less than 200 taxa in the backbone, group it as "Other"
import matplotlib.pyplot as plt
#plt.style.use('ggplot')
#remove background grid, set color to dark blue


gbif_backbone["phylum"] = gbif_backbone["phylum"].fillna("Unknown")
phylum_counts = gbif_backbone["phylum"].value_counts()
phylum_counts = phylum_counts.sort_values(ascending=False)
phylum_counts["Other"] = phylum_counts[phylum_counts <= 200].sum()
phylum_counts = phylum_counts[phylum_counts > 200]


phylum_counts.plot(kind='bar')
plt.xlabel("Phylum \n 64 total Phyla \n (Phyla with less than 200 taxa grouped as 'Other')")
# plt.show()
```
